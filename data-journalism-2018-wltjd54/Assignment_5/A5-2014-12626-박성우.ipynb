## 1. 미디어다음에서 제공하는 영어사전 웹사이트에서 랜돔으로 100개의 단어를 수집하고, 단어와 단어의 의미를 보기 좋게 출력하시오.

예: http://dic.daum.net/word/view.do?wordid=ekw000078303&q=help


* 저장 형식은 자유. 단, 가능한 한 보기 좋게 내용을 편집하여 저장하자.

  예)
  
    단어 1 
      1. 뜻풀이  
      2. 뜻풀이 
    ===  
    단어 2 
      1. 뜻풀이

```python
# 단어 URL을 살펴보면 http://dic.daum.net/word/view.do?wordid=ekw000388171의 형태로 되어 있고
# wordid를 바꿔주면 단어가 바뀐다. 그리고 현재 daum에서 제공하는 영어사전의 단어는 000000001부터 000388171, 총 388171개이다
from bs4 import BeautifulSoup
import urllib.request
import random as rd

# 1. random 수 형성하기
random_number = rd.sample(range(1, 388172), 100)
# 이때, word id로서 쓰이기 위해서는 0을 포함해 총 9자리여야 하며, 숫자가 없는 부분은 0으로 채워야 한다.
wordid_no = []
for number in random_number:
    id_number = "0" * (9 - len(str(number))) + str(number)
    wordid_no.append(id_number)

# 2. base_url 추출하기
base_url_list = ["http://dic.daum.net/word/view.do?wordid=ekw" + wordid for wordid in wordid_no]

# 3. html 불러오고 parsing하기 및 단어 출력하기
i = 0
for base_url in base_url_list:
    i += 1
    with urllib.request.urlopen(base_url) as url:
        doc = url.read()
    soup = BeautifulSoup(doc, "html.parser")
    words = soup.find_all("span", class_="txt_cleanword")[0] # 영어 단어 뽑아내기
    meanings = soup.find_all("span", class_="txt_mean") # 뜻 뽑아내기
    print(str(i) + '. %s' % words.text) # 단어 출력
    for j in range(len(meanings)):
        print(' ' * 5 + str(j + 1) + '. %s' % meanings[j].text) # 각 단어에 해당하는 뜻 출력
    if i < 100:
        print('\n=========\n')
    else:
        break
```
